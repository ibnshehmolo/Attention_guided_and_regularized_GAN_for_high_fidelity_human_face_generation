```python
"""
===============================================================================
ENHANCED DCGAN WITH LMHCSA ATTENTION
- Attention at 32×32 in Generator (64×64 → 128×128 path)
- Attention at 16×16 in Discriminator
- L2 Regularization (Weight Decay)
===============================================================================
"""

# ===================== AUTO-INSTALL =====================
import subprocess
import sys
import importlib

def install_if_missing(package):
    try:
        importlib.import_module(package)
    except ImportError:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package, "-q"])

install_if_missing("pyiqa")
install_if_missing("scikit-learn")
install_if_missing("tqdm")

# ===================== IMPORTS =====================
import os
import glob
import random
import warnings
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import torchvision
from torchvision import transforms, models
from torchvision.utils import save_image
from torch.utils.data import Dataset, DataLoader, Subset
from PIL import Image
import matplotlib.pyplot as plt
import pyiqa
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from scipy.linalg import sqrtm

warnings.filterwarnings("ignore")

# ===================== RANDOM SEED =====================
SEED = 42
torch.manual_seed(SEED)
torch.cuda.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
np.random.seed(SEED)
random.seed(SEED)

torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

print(f"Random seed fixed: {SEED}")

# ===================== DEVICE =====================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ===================== CONFIG =====================
CONFIG = {
    "DATASET_PATH": "/kaggle/input/datasets/gibi13/flickr-faces-hq-dataset-ffhq/images1024x1024",
    "IMG_SIZE": 128,                     # CHANGED: 64 → 128
    "Z_DIM": 256,
    "BATCH_SIZE": 16,
    "EPOCHS": 85,

    # SAME optimizer settings as proposed model
    "LR_G": 0.0002,
    "LR_D": 0.00008,

    # L2 Regularization (Weight Decay)
    "WEIGHT_DECAY_G": 0.0001,
    "WEIGHT_DECAY_D": 0.0,

    "TRAIN_RATIO": 0.75,
    "VAL_RATIO": 0.15,
    "TEST_RATIO": 0.10,

    "VAL_SUBSET_SIZE": 3000,

    "USE_SUBSET": True,
    "SUBSET_SIZE": 10000
}
os.makedirs("samples", exist_ok=True)
os.makedirs("final_results", exist_ok=True)

# ===================== LMHCSA ATTENTION MODULE =====================
class LMHCSA(nn.Module):
    """Local Multi-Head Channel Self-Attention"""
    def __init__(self, channels, num_heads=4):
        super().__init__()
        assert channels % num_heads == 0, "channels must be divisible by heads"
        
        self.num_heads = num_heads
        self.head_dim = channels // num_heads
        self.scale = self.head_dim ** -0.5

        self.qkv = nn.Conv2d(channels, channels * 3, 1)
        self.proj = nn.Conv2d(channels, channels, 1)
        self.gamma = nn.Parameter(torch.zeros(1))

    def forward(self, x):
        B, C, H, W = x.shape
        N = H * W

        qkv = self.qkv(x)
        q, k, v = torch.chunk(qkv, 3, dim=1)

        q = q.view(B, self.num_heads, self.head_dim, N).permute(0, 1, 3, 2)
        k = k.view(B, self.num_heads, self.head_dim, N).permute(0, 1, 3, 2)
        v = v.view(B, self.num_heads, self.head_dim, N).permute(0, 1, 3, 2)

        attn = torch.matmul(q, k.transpose(-2, -1)) * self.scale
        attn = torch.softmax(attn, dim=-1)

        out = torch.matmul(attn, v)
        out = out.permute(0, 1, 3, 2).contiguous().view(B, C, H, W)
        out = self.proj(out)

        return x + self.gamma * out

os.makedirs("samples", exist_ok=True)
os.makedirs("final_results", exist_ok=True)

# ===================== DATA LOADING =====================
all_files = glob.glob(os.path.join(CONFIG["DATASET_PATH"], "*.jpg")) + \
            glob.glob(os.path.join(CONFIG["DATASET_PATH"], "*.png"))

print(f"Total images found: {len(all_files)}")

if CONFIG["USE_SUBSET"]:
    random.seed(SEED)
    all_files = random.sample(all_files, CONFIG["SUBSET_SIZE"])
    print(f"Using subset: {len(all_files)}")

# ===================== SPLIT =====================
train_files, temp_files = train_test_split(
    all_files,
    test_size=(CONFIG["VAL_RATIO"] + CONFIG["TEST_RATIO"]),
    random_state=SEED
)

val_files, test_files = train_test_split(
    temp_files,
    test_size=CONFIG["TEST_RATIO"] / (CONFIG["VAL_RATIO"] + CONFIG["TEST_RATIO"]),
    random_state=SEED
)

print(f"Train: {len(train_files)}")
print(f"Val: {len(val_files)}")
print(f"Test: {len(test_files)}")

# ===================== DATASET =====================
class FaceDataset(Dataset):
    def __init__(self, files, augment=False):
        self.files = files

        if augment:
            self.transform = transforms.Compose([
                transforms.Resize((CONFIG["IMG_SIZE"], CONFIG["IMG_SIZE"])),
                transforms.RandomHorizontalFlip(0.5),
                transforms.ToTensor(),
                transforms.Normalize([0.5]*3, [0.5]*3)
            ])
        else:
            self.transform = transforms.Compose([
                transforms.Resize((CONFIG["IMG_SIZE"], CONFIG["IMG_SIZE"])),
                transforms.ToTensor(),
                transforms.Normalize([0.5]*3, [0.5]*3)
            ])

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        img = Image.open(self.files[idx]).convert("RGB")
        return self.transform(img)

train_dataset = FaceDataset(train_files, augment=True)
val_dataset = FaceDataset(val_files)
test_dataset = FaceDataset(test_files)

val_size = min(CONFIG["VAL_SUBSET_SIZE"], len(val_dataset))
val_indices = torch.randperm(len(val_dataset))[:val_size].tolist()
val_subset = Subset(val_dataset, val_indices)

train_loader = DataLoader(
    train_dataset,
    batch_size=CONFIG["BATCH_SIZE"],
    shuffle=True,
    drop_last=True
)

val_loader = DataLoader(
    val_subset,
    batch_size=CONFIG["BATCH_SIZE"],
    shuffle=False
)

test_loader = DataLoader(
    test_dataset,
    batch_size=CONFIG["BATCH_SIZE"],
    shuffle=False
)
# ===================== GENERATOR WITH ATTENTION AT 32×32 =====================
class Generator(nn.Module):
    def __init__(self):
        super().__init__()

        # Initial: 4×4
        self.initial = nn.Sequential(
            nn.ConvTranspose2d(CONFIG["Z_DIM"], 1024, 4, 1, 0, bias=False),
            nn.BatchNorm2d(1024),
            nn.ReLU(True),
        )

        # 4×4 → 8×8
        self.block1 = nn.Sequential(
            nn.ConvTranspose2d(1024, 512, 4, 2, 1, bias=False),
            nn.BatchNorm2d(512),
            nn.ReLU(True),
        )

        # 8×8 → 16×16
        self.block2 = nn.Sequential(
            nn.ConvTranspose2d(512, 256, 4, 2, 1, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU(True),
        )

        # 16×16 → 32×32
        self.block3 = nn.Sequential(
            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
        )

        # ⭐ ATTENTION AT 32×32 (128 channels) ⭐
        self.attention = LMHCSA(128, 4)

        # 32×32 → 64×64
        self.block4 = nn.Sequential(
            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
        )

        # 64×64 → 128×128
        self.block5 = nn.Sequential(
            nn.ConvTranspose2d(64, 32, 4, 2, 1, bias=False),
            nn.BatchNorm2d(32),
            nn.ReLU(True),
        )

        # Final output
        self.final = nn.Sequential(
            nn.Conv2d(32, 3, 3, padding=1),
            nn.Tanh()
        )

    def forward(self, z):
        z = z.view(z.size(0), CONFIG["Z_DIM"], 1, 1)
        x = self.initial(z)      # 1×1 → 4×4
        x = self.block1(x)       # 4×4 → 8×8
        x = self.block2(x)       # 8×8 → 16×16
        x = self.block3(x)       # 16×16 → 32×32
        x = self.attention(x)    # ⭐ Attention at 32×32
        x = self.block4(x)       # 32×32 → 64×64
        x = self.block5(x)       # 64×64 → 128×128
        return self.final(x)

class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()

        self.conv1 = nn.Sequential(
            nn.utils.spectral_norm(nn.Conv2d(3, 64, 4, 2, 1, bias=False)),
            nn.LeakyReLU(0.2, inplace=True),
        )

        self.conv2 = nn.Sequential(
            nn.utils.spectral_norm(nn.Conv2d(64, 128, 4, 2, 1, bias=False)),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),
        )

        self.conv3 = nn.Sequential(
            nn.utils.spectral_norm(nn.Conv2d(128, 256, 4, 2, 1, bias=False)),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2, inplace=True),
        )

        self.attention = LMHCSA(256, 4)

        self.conv4 = nn.Sequential(
            nn.utils.spectral_norm(nn.Conv2d(256, 512, 4, 2, 1, bias=False)),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.2, inplace=True),
        )

        self.conv5 = nn.Sequential(
            nn.utils.spectral_norm(nn.Conv2d(512, 1024, 4, 2, 1, bias=False)),
            nn.BatchNorm2d(1024),
            nn.LeakyReLU(0.2, inplace=True),
        )

        self.final = nn.utils.spectral_norm(
            nn.Conv2d(1024, 1, 4, 1, 0, bias=False)
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.attention(x)
        x = self.conv4(x)
        x = self.conv5(x)
        return self.final(x).view(-1, 1)

# ===================== WEIGHT INIT =====================
def weights_init(m):
    classname = m.__class__.__name__

    if classname.find("Conv") != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)

    elif classname.find("BatchNorm") != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)

# ===================== INIT MODELS =====================
G = Generator().to(device)
D = Discriminator().to(device)

G.apply(weights_init)
D.apply(weights_init)

# Count parameters
g_params = sum(p.numel() for p in G.parameters())
d_params = sum(p.numel() for p in D.parameters())
print(f"\nGenerator parameters: {g_params:,}")
print(f"Discriminator parameters: {d_params:,}")
print(f"Total parameters: {g_params + d_params:,}\n")

# ===================== OPTIMIZER WITH L2 REGULARIZATION =====================
optG = optim.Adam(
    G.parameters(),
    lr=CONFIG["LR_G"],
    betas=(0.5, 0.999),
    weight_decay=CONFIG["WEIGHT_DECAY_G"]  # L2 regularization
)

optD = optim.Adam(
    D.parameters(),
    lr=CONFIG["LR_D"],
    betas=(0.5, 0.999),
    weight_decay=CONFIG["WEIGHT_DECAY_D"]  # L2 regularization
)

criterion = nn.BCEWithLogitsLoss()

# ===================== FID =====================
class InceptionFeatureExtractor(nn.Module):
    def __init__(self):
        super().__init__()
        inception = models.inception_v3(weights=models.Inception_V3_Weights.DEFAULT)
        inception.fc = nn.Identity()
        self.model = inception.to(device).eval()

    def forward(self, x):
        x = (x + 1) / 2
        x = F.interpolate(x, size=(299, 299), mode='bilinear')
        return self.model(x)

feature_extractor = InceptionFeatureExtractor()

def compute_fid(loader, G, num_samples=5000):
    real_features = []
    fake_features = []
    count = 0

    with torch.no_grad():
        for real in loader:
            real = real.to(device)
            bs = real.size(0)

            z = torch.randn(bs, CONFIG["Z_DIM"]).to(device)
            fake = G(z)

            real_feat = feature_extractor(real).cpu().numpy()
            fake_feat = feature_extractor(fake).cpu().numpy()

            real_features.append(real_feat)
            fake_features.append(fake_feat)

            count += bs
            if count >= num_samples:
                break

    real_features = np.vstack(real_features)[:num_samples]
    fake_features = np.vstack(fake_features)[:num_samples]

    mu1 = real_features.mean(axis=0)
    mu2 = fake_features.mean(axis=0)

    sigma1 = np.cov(real_features, rowvar=False)
    sigma2 = np.cov(fake_features, rowvar=False)

    covmean = sqrtm(sigma1 @ sigma2)

    if np.iscomplexobj(covmean):
        covmean = covmean.real

    fid = np.sum((mu1 - mu2) ** 2) + np.trace(sigma1 + sigma2 - 2 * covmean)
    return float(fid)

# ===================== CW-SSIM =====================
cw_ssim_metric = pyiqa.create_metric("cw_ssim", device=device)

def compute_cwssim(real, fake):
    real = (real + 1) / 2
    fake = (fake + 1) / 2
    return cw_ssim_metric(real, fake).mean().item()

# ===================== TRACKING =====================
train_g_loss_hist = []
train_d_loss_hist = []
val_fid_hist = []
val_cwssim_hist = []

best_fid = float("inf")
fixed_z = torch.randn(16, CONFIG["Z_DIM"]).to(device)

# ===================== TRAINING =====================
print("\nStarting Enhanced DCGAN Training with LMHCSA Attention...\n")

for epoch in range(CONFIG["EPOCHS"]):

    G.train()
    D.train()

    epoch_g_loss = 0.0
    epoch_d_loss = 0.0
    num_batches = 0

    pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{CONFIG['EPOCHS']}")

    for real in pbar:
        real = real.to(device)
        bs = real.size(0)

        real_labels = torch.ones(bs, 1).to(device)
        fake_labels = torch.zeros(bs, 1).to(device)

        # ===================== TRAIN D =====================
        z = torch.randn(bs, CONFIG["Z_DIM"], device=device)
        fake = G(z)

        d_real = D(real)
        d_fake = D(fake.detach())

        d_loss = criterion(d_real, real_labels) + criterion(d_fake, fake_labels)

        optD.zero_grad()
        d_loss.backward()
        optD.step()

        # ===================== TRAIN G =====================
        output = D(fake)
        g_loss = criterion(output, real_labels)

        optG.zero_grad()
        g_loss.backward()
        optG.step()

        epoch_g_loss += g_loss.item()
        epoch_d_loss += d_loss.item()
        num_batches += 1

        pbar.set_postfix({
            "G": f"{g_loss.item():.3f}",
            "D": f"{d_loss.item():.3f}"
        })

    avg_g = epoch_g_loss / num_batches
    avg_d = epoch_d_loss / num_batches

    train_g_loss_hist.append(avg_g)
    train_d_loss_hist.append(avg_d)

    print(f"Epoch {epoch+1}: G={avg_g:.4f}, D={avg_d:.4f}")

    # ===================== VALIDATION (FIXED LOCATION) =====================
    if (epoch + 1) % 5 == 0:

        G.eval()
        D.eval()

        # -------- FID --------
        val_fid = compute_fid(val_loader, G)
        val_fid_hist.append(val_fid)

        # -------- CW-SSIM --------
        cw_scores = []

        with torch.no_grad():
            for real in val_loader:
                real = real.to(device)

                z = torch.randn(real.size(0), CONFIG["Z_DIM"], device=device)
                fake = G(z)

                cw_scores.append(compute_cwssim(real, fake))

        val_cw = sum(cw_scores) / len(cw_scores)
        val_cwssim_hist.append(val_cw)

        print(f"\nValidation -> FID: {val_fid:.2f}, CW-SSIM: {val_cw:.4f}")

        # -------- SAVE BEST --------
        if val_fid < best_fid:
            best_fid = val_fid
            torch.save(G.state_dict(), "best_enhanced_dcgan.pth")
            print("Best model saved.")

            with torch.no_grad():
                samples = G(fixed_z).cpu()

            save_image(samples, f"samples/epoch_{epoch+1}.png", nrow=4, normalize=True)

        G.train()
        D.train()

# ===================== TESTING =====================
print("\nTesting best model...\n")

G.load_state_dict(torch.load("best_enhanced_dcgan.pth"))
G.eval()

test_fid = compute_fid(test_loader, G)

test_cw_scores = []

with torch.no_grad():
    for real in test_loader:
        real = real.to(device)
        z = torch.randn(real.size(0), CONFIG["Z_DIM"]).to(device)
        fake = G(z)
        cw = compute_cwssim(real, fake)
        test_cw_scores.append(cw)

test_cwssim = np.mean(test_cw_scores)

print("\n================ FINAL RESULTS ================")
print(f"Test FID: {test_fid:.2f}")
print(f"Test CW-SSIM: {test_cwssim:.4f}")
print("==============================================")

# ===================== VISUALIZATION =====================
print("\nGenerating training curves...\n")

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Enhanced DCGAN (LMHCSA Attention) Training Curves", fontsize=16)

axs[0, 0].plot(train_g_loss_hist, linewidth=2)
axs[0, 0].set_title("Generator Loss")
axs[0, 0].set_xlabel("Epoch")
axs[0, 0].set_ylabel("Loss")
axs[0, 0].grid(True)

axs[0, 1].plot(train_d_loss_hist, linewidth=2)
axs[0, 1].set_title("Discriminator Loss")
axs[0, 1].set_xlabel("Epoch")
axs[0, 1].set_ylabel("Loss")
axs[0, 1].grid(True)

val_epochs = [(i + 1) * 5 for i in range(len(val_fid_hist))]
axs[1, 0].plot(val_epochs, val_fid_hist, marker='o', linewidth=2)
axs[1, 0].set_title("FID Progression")
axs[1, 0].set_xlabel("Epoch")
axs[1, 0].set_ylabel("FID")
axs[1, 0].grid(True)

axs[1, 1].plot(val_epochs, val_cwssim_hist, marker='o', linewidth=2)
axs[1, 1].set_title("CW-SSIM Progression")
axs[1, 1].set_xlabel("Epoch")
axs[1, 1].set_ylabel("CW-SSIM")
axs[1, 1].grid(True)

plt.tight_layout()
plt.savefig("enhanced_dcgan_training_curves.png", dpi=300)
plt.show()

# ===================== FINAL SAMPLES =====================
with torch.no_grad():
    final_samples = G(fixed_z).cpu()

grid = torchvision.utils.make_grid(final_samples, nrow=4, normalize=True)

plt.figure(figsize=(10, 10))
plt.imshow(grid.permute(1, 2, 0))
plt.axis("off")
plt.title(f"Enhanced DCGAN Samples (FID={best_fid:.2f})")
plt.savefig("final_enhanced_dcgan_samples.png", dpi=300)
plt.show()

print("\nTraining complete.")
print("Saved files:")
print("- best_enhanced_dcgan.pth")
print("- enhanced_dcgan_training_curves.png")
print("- final_enhanced_dcgan_samples.png")
print("- samples/")
```
